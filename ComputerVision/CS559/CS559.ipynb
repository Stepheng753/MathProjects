What is the storage saving in terms of $b$ and $c$ when reducing a b bit image into $c < b$ bit image?
Write a program to reduce an $b=8$ bit image into a $c<8$ bit image. Do not use the Matlab or Python library, but you can look their function/method to get ideas. Apply your code to the following 8-bit image, with $c=4$. Show the output image.

```python
def reduceImage(image, og_bit, new_bit):
    bit_diff = og_bit - new_bit
    diff_factor = np.power(2, bit_diff)

    image_og = plt.imread(image)

    image_new = (image_og / diff_factor) * diff_factor

    plt.figure()
    plt.imshow(image_og)
    plt.figure()
    plt.imshow(image_new)

reduceImage('./assets/hw1-3-1.jpg', 8, 4)
```
